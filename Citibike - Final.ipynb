```python
import pandas as pd
import json
from datetime import datetime

#files path
Jan_23_path='CSV Data/JC-202301-citibike-tripdata.csv'
Feb_23_path='CSV Data/JC-202302-citibike-tripdata.csv'
Mar_23_path='CSV Data/JC-202303-citibike-tripdata.csv'
Apr_23_path='CSV Data/JC-202304-citibike-tripdata.csv'
May_23_path='CSV Data/JC-202305-citibike-tripdata.csv'
Jun_23_path='CSV Data/JC-202306-citibike-tripdata.csv'
Jul_23_path='CSV Data/JC-202307-citibike-tripdata.csv'
Aug_23_path='CSV Data/JC-202308-citibike-tripdata.csv'
Sep_23_path='CSV Data/JC-202309-citibike-tripdata.csv'
Oct_23_path='CSV Data/JC-202310-citibike-tripdata.csv'
Nov_23_path='CSV Data/JC-202311-citibike-tripdata.csv'
Dec_23_path='CSV Data/JC-202312-citibike-tripdata.csv'


#Read data
Jan_23 = pd.read_csv(Jan_23_path)
Feb_23 = pd.read_csv(Feb_23_path)
Mar_23 = pd.read_csv(Mar_23_path)
Apr_23 = pd.read_csv(Apr_23_path)
May_23 = pd.read_csv(May_23_path)
Jun_23 = pd.read_csv(Jun_23_path)
Jul_23 = pd.read_csv(Jul_23_path)
Aug_23 = pd.read_csv(Aug_23_path)
Sep_23 = pd.read_csv(Sep_23_path)
Oct_23 = pd.read_csv(Oct_23_path)
Nov_23 = pd.read_csv(Nov_23_path)
Dec_23 = pd.read_csv(Dec_23_path)


#Combine data to a single dataframe
rides_df=pd.concat([Jan_23, Feb_23, Mar_23, Apr_23, May_23, Jun_23, Jul_23, Aug_23, Sep_23, Oct_23, Nov_23, Dec_23],ignore_index=True)

#Display data table for preview
rides_df
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime
rides_df['started_at'] = pd.to_datetime(rides_df['started_at'])
rides_df['ended_at'] = pd.to_datetime(rides_df['ended_at'])
```

```python
# Ride Analysis
# Extract month from 'started_at' column
rides_df['month'] = rides_df['started_at'].dt.month_name()

# Total rides by month
total_rides_by_month = rides_df['month'].value_counts().sort_index().astype(int)

# Total rides by member type by month
total_rides_by_member_type_by_month = rides_df.groupby(['month', 'member_casual']).size().unstack(fill_value=0).astype(int).reset_index()
total_rides_by_member_type_by_month = total_rides_by_member_type_by_month.to_dict(orient='records')

# Total rides by rideable type by month
total_rides_by_rideable_type_by_month = rides_df.groupby(['month', 'rideable_type']).size().unstack(fill_value=0).astype(int).reset_index()
total_rides_by_rideable_type_by_month = total_rides_by_rideable_type_by_month.to_dict(orient='records')
```

```python
# Prepare the final JSON output
output_data = {
    'total_rides_by_month': total_rides_by_month.reset_index().to_dict(orient='records'),
    'total_rides_by_member_type_by_month': total_rides_by_member_type_by_month,
    'total_rides_by_rideable_type_by_month': total_rides_by_rideable_type_by_month
}
```

```python
# Save the combined dataset to a single JSON file
with open('ride_analysis_by_month.json', 'w') as f:
    json.dump(output_data, f)
```

```python
#List all start stations

# Group by start station name and coordinates, then count rides
start_station_coordinates = rides_df.groupby(['start_station_name', 'start_lat', 'start_lng']).size().reset_index(name='count')

# Find the most frequently occurring latitude and longitude for each end station
most_used_start_coordinates = start_station_coordinates.sort_values('count', ascending=False).drop_duplicates(['start_station_name'])

# Rename columns to avoid conflicts during merge
most_used_coordinates.rename(columns={'start_lat': 'most_used_start_lat', 'start_lng': 'most_used_start_lng'}, inplace=True)

# Ensure no conflicting column names before the merge
if 'most_used_start_lat' in start_station_rides.columns or 'most_used_start_lng' in start_station_rides.columns:
    start_station_rides.drop(columns=['most_used_start_lat', 'most_used_start_lng'], inplace=True)

# Merge with start station rides data to associate each station with its most used coordinates
start_station_rides = pd.merge(start_station_rides, most_used_coordinates[['start_station_name', 'most_used_start_lat', 'most_used_start_lng']], on='start_station_name', how='left')

# Convert to GeoJSON format
start_station_rides_geojson_output = {
    "type": "FeatureCollection",
    "features": []
}

for _, row in start_station_rides.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row['most_used_start_lng'], row['most_used_start_lat']]
        },
        "properties": {
            "name": row['start_station_name'],
            "ride_count": row['ride_count']
        }
    }
    start_station_rides_geojson_output['features'].append(feature)

# Save GeoJSON output to file
with open('start_station_rides.geojson', 'w') as f:
    json.dump(start_station_rides_geojson_output, f)
```

```python
# Group by end station name and coordinates, then count rides
end_station_coordinates = rides_df.groupby(['end_station_name', 'end_lat', 'end_lng']).size().reset_index(name='count')

# Find the most frequently occurring latitude and longitude for each end station
most_used_end_coordinates = end_station_coordinates.sort_values('count', ascending=False).drop_duplicates(['end_station_name'])

# Rename columns to avoid conflicts during merge
most_used_end_coordinates.rename(columns={'end_lat': 'most_used_end_lat', 'end_lng': 'most_used_end_lng'}, inplace=True)

# Ensure no conflicting column names before the merge
if 'most_used_end_lat' in end_station_rides.columns or 'most_used_end_lng' in end_station_rides.columns:
    end_station_rides.drop(columns=['most_used_end_lat', 'most_used_end_lng'], inplace=True)

# Merge with end station rides data to associate each station with its most used coordinates
end_station_rides = pd.merge(end_station_rides, most_used_end_coordinates[['end_station_name', 'most_used_end_lat', 'most_used_end_lng']], on='end_station_name', how='left')

# Convert to GeoJSON format
end_station_rides_geojson_output = {
    "type": "FeatureCollection",
    "features": []
}

for _, row in end_station_rides.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row['most_used_end_lng'], row['most_used_end_lat']]
        },
        "properties": {
            "name": row['end_station_name'],
            "ride_count": row['ride_count']
        }
    }
    end_station_rides_geojson_output['features'].append(feature)

# Save GeoJSON output to file
with open('end_station_rides.geojson', 'w') as f:
    json.dump(end_station_rides_geojson_output, f)

```
